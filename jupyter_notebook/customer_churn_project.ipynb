```python
import pandas as pd
df=pd.read_csv(r"C:\Users\DELL\OneDrive\Documents\6SEM INTERNSHIP PROJECT\anurag project antivirus\Telco-Customer-Churn.csv")
print("Shape of dataset:",df.shape)
print("Column Types:\n",df.dtypes)
df.head()
```

```python
df['TotalCharges'].isna().sum()
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
df['TotalCharges'].isna().sum()
```

```python
df['TotalCharges'].fillna(df['TotalCharges'].median(), inplace=True)
df['TotalCharges'].isna().sum()
df['TotalCharges'].dtype
```

```python
df['Churn'].head()
```

```python
df['Churn'] = df['Churn'].map({'Yes': 1, 'No': 0})
```

```python
df['Churn'].head()
```

```python
df['TenureGroup'] = pd.cut(
    df['tenure'],
    bins=[0, 12, 24, 48, 60, 72],
    labels=['0-1 yr', '1-2 yrs', '2-4 yrs', '4-5 yrs', '5+ yrs']
)
df['HighCharges'] = df['MonthlyCharges'].apply(lambda x: 1 if x > df['MonthlyCharges'].median() else 0)
```

```python
df.to_csv(r"C:\Users\DELL\OneDrive\Documents\6SEM INTERNSHIP PROJECT\anurag project antivirus\data_preprocessed_telco_churn.csv", index=False)
```
